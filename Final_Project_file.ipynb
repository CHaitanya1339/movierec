```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity

# Convert bag of words to TF-IDF  
tfidf_transformer = TfidfTransformer()
tfidf = tfidf_transformer.fit_transform(vec_matrix)

# Apply LSA or LSI
lsa = TruncatedSVD(n_components=20, algorithm='arpack')
lsa.fit(tfidf)

original_data = tfidf.toarray()
reconstructed_data = lsa.inverse_transform(lsa.transform(original_data))
reconstruction_error = np.sqrt(np.mean((original_data - reconstructed_data)**2))
print("Reconstruction Error:",reconstruction_error)
```

```python
from tabulate import tabulate
from fuzzywuzzy import process
import imdb
from fuzzywuzzy import fuzz


import requests
import time

def get_streaming_details(movie_title):
    url = "https://ott-details.p.rapidapi.com/search"
    querystring = {"title": movie_title, "page": "1"}

    headers = {
      "X-RapidAPI-Key": "88c04308e6msh92b978067113888p121c5ejsn5647af504e9e",
      "X-RapidAPI-Host": "ott-details.p.rapidapi.com"
    }


    response = requests.request("GET", url, headers=headers, params=querystring)

    # Convert response to JSON
    json_response = response.json()
    imdb_id = None
    for result in json_response['results']: 
      if result['title'] == querystring['title']:
          imdb_id = result['imdbid']
          break
    if imdb_id is None:
        return None
    time.sleep(1)

    url = "https://ott-details.p.rapidapi.com/gettitleDetails"
    querystring = {"imdbid": imdb_id}
    
    response = requests.request("GET", url, headers=headers, params=querystring)
    json_response = response.json()

    streaming_details = {}
    if 'streamingAvailability' not in json_response or not json_response['streamingAvailability']:
        return None
    streaming_availability = json_response['streamingAvailability']['country']
    for country, streaming_info in streaming_availability.items():
        streaming_details[country] = []
        for info in streaming_info:
            platform = info['platform']
            url = info['url']
            streaming_details[country].append({
                "platform": platform,
                "url": url
            })

    return streaming_details 



# Initialize the IMDb client
ia = imdb.IMDb()

# Get the user input using fuzzy matching
user_movie = input("Enter a movie title: ")
movie_title, movie_score, movie_index = process.extractOne(user_movie, dataset['title'], scorer=fuzz.token_sort_ratio)

if movie_score < 90:
    print("Sorry, we could not find a close match for your movie title.")
else:
    print(f"Movie selected: {movie_title}")

    # Find the index of the user movie
    movie_index = dataset[dataset['title'] == movie_title].index[0]

    # Compute the cosine similarities between the user movie and all other movies
    similarity_scores = cosine_similarity(tfidf[movie_index], tfidf)

    # Get the top 10 most similar movies
    similar_movies = list(enumerate(similarity_scores[0]))
    sorted_similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)[1:50]

    # Ask the user for language preferences
    language_preferences = input("Enter your preferred languages separated by commas: ").lower().split(",")
    
    print("Recommending...")


    # Create a list of tuples with the movie data
    movie_data = []
    for i, score in sorted_similar_movies:
        title = dataset.loc[i, 'title']
        year = dataset.loc[i, 'year']
        rating = dataset.loc[i, 'rating']
        certificate = dataset.loc[i, 'certificate']
        
        # Skip adding the user input movie title to the list
        if title == movie_title:
          continue
       
        search_query = f"{title} ({year})"
        search_results = ia.search_movie(search_query)

        if search_results:
            movie = ia.get_movie(search_results[0].getID())
            cast = movie.get('cast')
            if cast:
                cast = [person['name'] for person in cast][:5]
                cast_str=",".join(cast)
            else:
                cast_str = "Unknown"
            directors = movie.get('director')
            if directors:
                directors = [person['name'] for person in directors][:3]
                directors_str = ", ".join(directors)
            else:
                directors_str = "Unknown"
            language = movie.get('language', ['Unknown'])
            language_str = ", ".join(language)
            for lang in language_preferences:
                if lang in language_str.lower():
                    streaming_details = get_streaming_details(title)
                    if streaming_details is None:
                        streaming_details_str = "Not Available"
                    else:
                        streaming_details_str = str(streaming_details)
                    movie_data.append((title, year, rating, certificate, language_str, directors_str, cast_str, streaming_details_str))
                    break
        else:
            print(f"No search results found for {title} ({year})")


        if len(movie_data) == 10:
            break

    # Print the movie data in a table format
    headers = ['Title', 'Year', 'Rating', 'Certificate', 'Language', 'Director', 'Cast', 'Streaming Details']
    print(tabulate(movie_data, headers=headers, tablefmt='grid'))
```
