```python
dataset['movie_title'] = dataset['movie_title'].apply(lambda a:a[:-1])
dataset['movie_title'][0]
```

```python
## combined features on which we will calculate cosine similarity

dataset['director_genre_actors'] = dataset['director_name']+' '+dataset['actor_1_name']+' '+dataset['actor_2_name']+' '+dataset['actor_3_name']+' '+dataset['genres']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vec = CountVectorizer()
vec_matrix = vec.fit_transform(dataset['director_genre_actors'])
print(vec_matrix)



similarity = cosine_similarity(vec_matrix)
```

Using Latent Semantic Analysis

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.decomposition import TruncatedSVD
# Convert bag of words to TF-IDF
tfidf_transformer = TfidfTransformer()
tfidf = tfidf_transformer.fit_transform(vec_matrix)

# Apply LSA or LSI
lsa = TruncatedSVD(n_components=100, algorithm='arpack')
lsa.fit(tfidf)
```

Recommending Using   Latent Semantic Analysis

```python
# Get the user input
user_movie = input("Enter a movie title: ")

# Find the index of the user movie
movie_index = dataset[dataset['movie_title'] == user_movie].index[0]

# Compute the cosine similarities between the user movie and all other movies
similarity_scores = cosine_similarity(tfidf[movie_index], tfidf)
similarity_scores

# Get the top 10 most similar movies
similar_movies = list(enumerate(similarity_scores[0]))
sorted_similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)[1:20]
# Print the top 10 similar movies
for i, score in sorted_similar_movies:
   print("{}: {}".format(i, df.loc[i, 'movie_title']))
```

**Recommendation Using Binary Feature Matrix Vectorizer**

```python
movie=input("Enter a Movie Name-")
year=int(input("Enter the minimum released year-"))
min_age=int(input("Enter the minimum age who wants to watch-"))
allowed_rating = []
if (min_age < 7):
    allowed_rating = ["TV-Y"]
elif min_age <= 13:
    allowed_rating = ["TV-Y", "TV-Y7", "TV-Y7-FV"]
elif min_age <= 14:
    allowed_rating = ["TV-Y", "TV-Y7", "TV-Y7-FV", "PG-13", "TV-14"]
else:
    allowed_rating = ["TV-Y", "TV-Y7", "TV-Y7-FV", "PG-13", "TV-PG", "TV-14", "R", "TV-MA", "NR", "PG", "UR", "NC-17"]

if movie not in dataset['movie_title'].unique():
    print('Sorry! The movie you requested is not in our database. Please check the spelling or try with some other movies')
else:
    i = dataset.loc[dataset['movie_title']==movie].index[0]
    lst = list(enumerate(similarity[i]))
    lst = sorted(lst, key=lambda x: x[1], reverse=True)
    lst = lst[1:11] # excluding first item since it is the requested movie itself
    l = []
    year_list = []
    for i in range(len(lst)):
        a = lst[i][0]
        if dataset['title_year'][a] > year and dataset['content_rating'][a] in allowed_rating:
            l.append(dataset['movie_title'][a])
            year_list.append(dataset['title_year'][a])
    if not l:
        print('No recommended movies found after ' + str(year) + ' with the selected age rating')
    plt.figure(figsize=(10, 5))
    plt.bar(l, [i[1] for i in lst if dataset['title_year'][i[0]] > year and dataset['content_rating'][i[0]] in allowed_rating])
    plt.xticks(rotation=90)
    plt.xlabel('Movies similar to â†’ ' + movie+" using Binary Feature Matrix Vectorizer", fontsize=12, fontweight="bold")
    plt.ylabel('cosine scores', fontsize=12, fontweight="bold")
    plt.show()
    df2 = pd.DataFrame({'Movies Recommended': l, 'Year': year_list})
    df2.drop_duplicates(inplace=True)
    print(df2)
```

**Recommendation Using LSA**

```python
movie=input("Enter a Movie Name-")
year=int(input("Enter the minimum released year-"))
min_age=int(input("Enter the minimum age who wants to watch-"))
movie_index = dataset[dataset['movie_title'] == movie].index[0]
allowed_rating = []
if (min_age < 7):
    allowed_rating = ["TV-Y"]
elif min_age <= 13:
    allowed_rating = ["TV-Y", "TV-Y7", "TV-Y7-FV"]
elif min_age <= 14:
    allowed_rating = ["TV-Y", "TV-Y7", "TV-Y7-FV", "PG-13", "TV-14"]
else:
    allowed_rating = ["TV-Y", "TV-Y7", "TV-Y7-FV", "PG-13", "TV-PG", "TV-14", "R", "TV-MA", "NR", "PG", "UR", "NC-17"]
similarity_scores = cosine_similarity(tfidf[movie_index], tfidf)

similar_movies = list(enumerate(similarity_scores[0]))
sorted_similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)[1:11]
l = []
year_list = []
for movie_index in range(len(sorted_similar_movies)):
    a = sorted_similar_movies[movie_index][0]
    if dataset['title_year'][a] > year and dataset['content_rating'][a] in allowed_rating:
        l.append(dataset['movie_title'][a])
        year_list.append(dataset['title_year'][a])
if not l:
    print('No recommended movies found after ' + str(year) + ' with the selected age rating')
plt.figure(figsize=(10, 5))
plt.bar(l, [movie_index[1] for movie_index in sorted_similar_movies if dataset['title_year'][movie_index[0]] > year and dataset['content_rating'][movie_index[0]] in allowed_rating])
plt.xticks(rotation=90)
plt.xlabel('Movies similar to â†’ ' + movie+' using LSA', fontsize=12, fontweight="bold")
plt.ylabel('cosine scores', fontsize=12, fontweight="bold")
plt.show()
df3 = pd.DataFrame({'Movies Recommended': l, 'Year': year_list})
df3.drop_duplicates(inplace=True)
print(df3)
```
